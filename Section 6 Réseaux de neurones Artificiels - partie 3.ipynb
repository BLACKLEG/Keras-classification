```python
#set de feat / target
X = df.drop ( 'benign_0__mal_1' , axis = 1).values
y = df.benign_0__mal_1.values
```

```python
# split train test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, 
                                                    test_size=0.25, 
                                                    random_state=101)
```

```python
# normalisation

from sklearn.preprocessing import MinMaxScaler 

scaler = MinMaxScaler ()

X_train = scaler.fit_transform ( X_train )

X_test = scaler.transform ( X_test )
```

```python
X_train.shape
```

```python
# reseau de neurone
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# reseau de neurone
model = Sequential ()
model.add (Dense (30, activation='relu') )
model.add (Dense (15, activation='relu') )
model.add (Dense (1, activation='sigmoid') )  # c est une classification 
# dans ce cas la il ne faut pas utiliser relu (regression)
# sigmoid retourne une valeur 0 ou 1

model.compile ( 'Adam', 
               loss = 'binary_crossentropy' )
# cest une classification donc la fct de perte doit 
# permettre de determinÃ© l erreur pour ce type d application

model.fit ( x=X_train,
            y=y_train,
            epochs=600,
            validation_data= (X_test , y_test) )
```

```python
#tracer loss evolution
pd.DataFrame(model.history.history).plot()

#ici c est un cas de surapprentissage 
#plus il y a d epochs plus la loss du set de train est meilleur et 
# le set de validation est mauvais
```

```python
# utilisation d early stopping

from tensorflow.keras.callbacks import EarlyStopping

#callbacks 
early_stop = EarlyStopping (monitor = 'val_loss' ,   #valeur monitorÃ©e : val_los
                            mode = 'min' ,   #minimisÃ© la val loss
                            verbose = 1 ,   # rapport
                            patience = 25 )  # nombre d epoch apres detection
#reseau de neurone
model = Sequential ()
model.add (Dense (30, activation='relu') )
model.add (Dense (15, activation='relu') )
model.add (Dense (1, activation='sigmoid') )  # c est une classification 

model.compile ( 'Adam', 
               loss = 'binary_crossentropy' )

model.fit ( x=X_train,
            y=y_train,
            epochs=600,
            validation_data= (X_test , y_test) ,
            callbacks = (early_stop))
```

```python
#tracer loss evolution
pd.DataFrame(model.history.history).plot()

#precedament la courbe apprentissage partait en cas de surapprentissage 
# l utilisation du callbacks permet d eviter que cela arrive 
# et arrete prematurement l apprentissage 
```

```python
from tensorflow.keras.layers import Dropout


#reseau de neurone
model = Sequential ()
model.add (Dense (30, activation='relu') )
model.add (Dropout (rate = 0.5) ) # la moitiÃ© des neuronnes (30) de la couche 
                                  # precedente sont desactivÃ©s aleatoirement
model.add (Dense (15, activation='relu') )
model.add (Dropout (rate = 0.5) ) # la moitiÃ© des 15 neuronnes de la couche 
model.add (Dense (1, activation='sigmoid') )  # c est une classification 

#compilateur
model.compile ( 'Adam', 
               loss = 'binary_crossentropy' )

#callbacks 
early_stop = EarlyStopping (monitor = 'val_loss' ,   #valeur monitorÃ©e : val_los
                            mode = 'min' ,   #minimisÃ© la val loss
                            verbose = 1 ,   # rapport
                            patience = 25 )  # nombre d epoch apres detection

#entrainement
model.fit ( x=X_train,
            y=y_train,
            epochs=600,
            validation_data= (X_test , y_test) ,
            callbacks = (early_stop))
```

```python
#tracer loss evolution
pd.DataFrame(model.history.history).plot()

# la performence est significativement ameliorer
```

```python
#estimer la qualitÃ© du model avec des metrics d erreurs

pred = (model.predict ( X_test ) > 0.5).astype ('int32')   #permet de binariser les predictions
pred.shape
#prediction
pred = pd.Series (pred.reshape (143, ) )
real = pd.Series (y_test )

loss_df = pd.concat ( [pred , real] , axis = 1)
loss_df.rename ( columns = { 0 : "pred" , 1 : 'real' } , inplace = True )
loss_df

from sklearn.metrics import classification_report , confusion_matrix

print (classification_report ( loss_df.pred  , loss_df.real ))

print ()


print(confusion_matrix ( loss_df.pred  , loss_df.real  ))

#nous avons une bonne performance
# avec une bonne feature enginering la performance peut etre obtimisÃ©
```
